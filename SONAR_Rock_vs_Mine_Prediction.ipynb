```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing

```python
Sonar_data=pd.read_csv('/content/Copy of sonar data.csv')
```

```python
Sonar_data.head()
```

```python
Sonar_data.iloc[:, -1].value_counts()
```

```python
# separating into features and target
X = Sonar_data.drop(columns=Sonar_data.columns[-1], axis=1)
Y = Sonar_data[Sonar_data.columns[-1]]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_test.shape)
```

Use the logistic regression

```python
#train the logistic regression model with the data
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
```

```python
#accuarcy on training data

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data

X_test_prediction=model.predict(X_test)
training_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("Accuracy on test data : ",training_data_accuracy)
```
